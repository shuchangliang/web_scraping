```python
import requests
from requests import get
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
# define the url that you are scraping
url = 'http://itianti.onerank.cn/index/cpu/index.html'
```

```python
# query the website and return the html to the variable 'page'
page = urllib.request.urlopen(url)

# parse the html using beautiful soup and store in variable 'soup'
soup = BeautifulSoup(page, 'html.parser')
print(soup.prettify())
```

#### Here is the information we’ll gather from each movie listing:
- Rank
- CPU Name
- Score

```python
# create and write headers to a list 
rows = []
rows.append(['Rank', 'CPU Name', 'Score'])
print(rows)
```

```python
# find results within table
table = soup.find('table', attrs={'id': 'maintable'})
results = table.find_all('tr')
print('Number of results', len(results))
```

```python
# loop over results
for result in results:
    # find all columns per result
    data = result.find_all('td')
    # check that columns have data 
    if len(data) == 0: 
        continue
    
    # write columns to variables
    rank = data[0].getText()
    cpuname = data[1].getText()
    score = data[3].getText()
    
    # remove unwanted characters
    cpuname = cpuname.strip("\n")

    # write each result to rows
    rows.append([rank, cpuname, score])

print(rows)
```

```python
df = pd.DataFrame(rows)
df
```

```python
df.columns = ['Rank', 'CPU Name', 'Score']
df.drop(df.index[0], inplace=True)
df.set_index('Rank', inplace=True)
```
